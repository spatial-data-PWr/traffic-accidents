```python
import sys
import folium
import numpy as np
import pickle as pkl

sys.path.append('..')

from pyproj import Transformer
from src import STORAGE_PATH
```

### Load graphs

```python
DATA_DIRECTORY = STORAGE_PATH / 'graphs'

with open(DATA_DIRECTORY / 'bike_accidents_graph.pkl', 'rb') as f:
    bike_accidents_graph = pkl.load(f)

with open(DATA_DIRECTORY / 'nonbike_accidents_graph.pkl', 'rb') as f:
    nonbike_accidents_graph = pkl.load(f)



with open(DATA_DIRECTORY / 'bikepaths.pkl', 'rb') as f:
    bikepaths = pkl.load(f)

with open(DATA_DIRECTORY / 'osm_roads_graph.pkl', 'rb') as f:
    osm_roads_graph = pkl.load(f)
```

```python
print(f'Bike accidents count: {len(bike_accidents_graph)}')
print(f'Nonbike accidents count: {len(nonbike_accidents_graph)}')
```

```python
folium_map = folium.Map() #bikepaths.explore()

in_proj = 'epsg:2177'
out_proj = 'epsg:4326'
subsample_rate = 0.05
transformer = Transformer.from_crs(in_proj, out_proj)

for node in np.random.permutation(list(bike_accidents_graph))[0:int(subsample_rate * len(bike_accidents_graph))]:
    folium.Marker(transformer.transform(node[1], node[0]), icon=folium.Icon(color="red", icon="info-sign")).add_to(folium_map)

for node in np.random.permutation(list(nonbike_accidents_graph))[0:int(subsample_rate * len(nonbike_accidents_graph))]:
    folium.Marker(transformer.transform(node[1], node[0]), icon=folium.Icon(color="blue", icon="info-sign")).add_to(folium_map)
```
